```python
import pandas as pd
df = pd.read_csv('/Users/patrickcullinane/Desktop/data/train.csv')
df.columns = [c.lower() for c in df.columns] #postgres doesn't like capitals or spaces

from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:cull2@CULL@localhost:5432/Molecule')

df.to_sql("data_table", engine)
```

```python
query = "SELECT molecule_name, COUNT(1) OCCURENCES \
            FROM data_table \
            GROUP BY molecule_name \
            ORDER BY OCCURENCES ASC"
df_4 = pd.read_sql_query(query, engine); df_4.head()
```

```python
query = "SELECT molecule_name, type, COUNT(1) OCCURENCES \
            FROM data_table \
            WHERE type = '1JHC' \
            GROUP BY type, molecule_name \
            ORDER BY OCCURENCES ASC"
df_3 = pd.read_sql_query(query, engine); df_3
```
